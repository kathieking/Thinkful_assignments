```python
fig = plt.figure()
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

color = ['cyan', 'black', 'magenta', 'yellow', 'green', 'purple', 'red', 'blue']
#
        

ax1 = plt.subplot(2, 2, 1)
df_2009.plot(ax = ax1, kind = 'bar', x = 'Area', y = 'Homeless Estimates', color = color, legend = None)
plt.ylabel('Total Homeless')
plt.title('2009 New York City Homeless Population by Area')

ax2 = plt.subplot(2, 2, 2)
df_2010.plot(ax = ax2, kind = 'bar', x = 'Area', y = 'Homeless Estimates', color = color, legend = None)
plt.ylabel('Total Homeless')
plt.title('2010 New York City Homeless Population by Area')

ax3 = plt.subplot(2, 2, 3)
df_2011.plot(ax = ax3, kind = 'bar', x = 'Area',  y = 'Homeless Estimates', color = color, legend = None)
plt.ylabel('Total Homeless')
plt.title('2011 New York City Homeless Population by Area')

ax4 = plt.subplot(2, 2, 4)
df_2012.plot(ax = ax4, kind = 'bar', x = 'Area', y = 'Homeless Estimates', color = color, legend = None)
plt.ylabel('Total Homeless')
plt.title('2012 New York City Homeless Population by Area')

plt.tight_layout()
plt.show()
```

Plot 2 Trends: 

> The homeless population in each borough appears to be relativley consistent over a 4-year period.

>The areas with the highest homeless populations are clearly the Subways and Manhattan, followed by Brooklyn. 

>The remaining areas appear to have a significantly lower homeless population, perhaps a reflection of lower overall population? 

>The incidence of homeless in the Subways is most interesting, as it is much less visible to a casual observer. 

>The trend appears to show a decrease for each area in year 2011, before increasing to its highest point in 2012. 

>Recommend further investigation into cause related to decrease in 2011.

# Plot 3: New York City Homelss Population by Area over 4-year period

```python
df['order'] = df['Year'].map({2009: 0, 2010: 1, 2011: 2, 2012: 3})
```

```python
df.sort_values('Year', ascending = True, inplace= True)
```

```python
df.sort_values(by = ['Area', 'Year'], ascending = True, inplace= True)
```

```python
fig, axes = plt.subplots(4,2, figsize=(15,5))
for (area, group), ax in zip(df.groupby('Area'), axes.flatten()):
    group.plot(x='Year', y='Homeless Estimates', kind='line', legend=None, ax=ax, xticks =([2009,2010,2011,2012]),title= area)
    
plt.show()
```

```python
df_sub = df[df['Area'] == 'Subways ']
df_brx = df[df['Area'] == 'Surface Area - Bronx ']
df_brk = df[df['Area'] == 'Surface Area - Brooklyn ']
df_man = df[df['Area'] == 'Surface Area - Manhattan ']
df_qns = df[df['Area'] == 'Surface Area - Queens ']
df_sti = df[df['Area'] == 'Surface Area - Staten Island ']
df_surt = df[df['Area'] == 'Surface Total ']
df_tot = df[df['Area'] == 'Total Unsheltered Individuals ']
```

```python
df_sub.head()
```

```python
fig = plt.figure()
fig, ax = plt.subplots(nrows=4, ncols=2, figsize=(10, 10))


ax1 = plt.subplot(4, 2, 1)
df_sub.plot(ax  = ax1, kind = 'line', x = 'Year',y = 'Homeless Estimates',  title = 'New York City Homeless Population: Subway', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

ax2 = plt.subplot(4, 2, 2)
df_brx.plot(ax  = ax2, kind = 'line', x = 'Year',y = 'Homeless Estimates',  title = 'New York City Homeless Population: Bronx', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

ax3 = plt.subplot(4, 2, 3)
df_brk.plot(ax  = ax3, kind = 'line', x = 'Year',y = 'Homeless Estimates',  title = 'New York City Homeless Population: Brooklyn', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

ax4 = plt.subplot(4, 2, 4)
df_man.plot(ax  = ax4, kind = 'line', x = 'Year',y = 'Homeless Estimates',  title = 'New York City Homeless Population: Manhattan', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

ax5 = plt.subplot(4, 2, 5)
df_qns.plot(ax  = ax5, kind = 'line', x = 'Year',y = 'Homeless Estimates',  title = 'New York City Homeless Population: Queens', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

ax6 = plt.subplot(4, 2, 6)
df_sti.plot(ax  = ax6, kind = 'line', x = 'Year',y = 'Homeless Estimates',  title = 'New York City Homeless Population: Staten Island', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

ax7 = plt.subplot(4, 2, 7)
df_surt.plot(ax  = ax7, kind = 'line', x = 'Year',y = 'Homeless Estimates',  title = 'New York City Homeless Population: Surface Total', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

ax8 = plt.subplot(4, 2, 8)
df_tot.plot(ax  = ax8, kind = 'line', x = 'Year',y = 'Homeless Estimates', title = 'New York City Homeless Population: Total', legend = None)
plt.xticks([2009, 2010, 2011, 2012])

plt.tight_layout()
```

Plot 3 Trends: 

> The homeless population in the Subways indicates a consistent upward trend over a 4-year period.

> The incidence of homeless in the Subways is most interesting, as it is much less visible to a casual observer. 

> The trend appears to show a decrease for each area in year 2011, except for the Subways. 

> Both Staten Island and Queens show a consistent downward trend in the homeless population after 2010, resulting in an overall decrease in homeless population. 

> Bronx, Brooklyn and Manhattan, as well as the Surface Total show an increase after the 2011 dip. 

> Bronx population appears relatively stable over a 4 year period. 

> Brooklyn, Manhattan, and the Surface Total show an overall increase in the homeless population over the 4-year period. 

> Subways and the Total Unsheltered Individuals plots also show an overall increase over the 4-year period. It is possible that the areas which show on overall decrease in homeless population may reflect a transfer of the homeless to the Subways rather than an actual decrease in the overall indicence of homelessness.

>Further investigation of the factors contributing to decreased or steady homeless populations in some areas while other areas show varying degress of increase is warranted.

# Plot 4: Changes in New York City Homeless Popultaion over 4-year period

```python
table = pd.pivot_table(df, values='Homeless Estimates', index=['Area'], columns=['Year'], aggfunc = 'mean')
```

```python
table['Change in Population'] = table[2012] - table[2009]
```

```python
table['Change in Population'] = table[2012] - table[2009]
```

```python
table.plot(kind = 'bar', y = 'Change in Population', legend = None)
plt.ylabel('Change in Population')
plt.title('Changes in New York City Homeless Popultaion over 4-year period')
plt.show()
```

Plot 4 trends:

It is possible that the areas which show on overall decrease in homeless population may reflect a transfer of the homeless to the Subways rather than an actual decrease in the overall indicence of homelessness.
